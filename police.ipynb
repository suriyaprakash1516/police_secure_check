```python
# convert stop date to correct format(YYYY-MM-DD)
df['stop_date']=pd.to_datetime(df['stop_date'],errors='coerce').dt.strftime('%Y-%m-%d')
df['stop_time']=pd.to_datetime(df['stop_time'],errors='coerce').dt.strftime('%H:%M:%S')
```

```python
# prepare insert query
insert_query="""
INSERT INTO police_logs(	
     stop_date,stop_time,country_name,driver_gender,driver_age,
     driver_race,violation,search_conducted,search_type,stop_outcome,
     drugs_related_stop,stop_duration,vehicle_number,timestamp
     )VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
"""
```

```python
import pandas as pd
import numpy as np

# convert ALL NaN → None (IMPORTANT)
df = df.replace({np.nan: None})

def to_int(x):
    return int(x) if x is not None else None

def to_bool(x):
    if x is None:
        return None
    return 1 if str(x).strip().lower() in ["true", "1", "yes"] else 0

# Insert
for idx, row in df.iterrows():
    values = (
        row.get('stop_date'),
        row.get('stop_time'),
        row.get('country_name'),
        row.get('driver_gender'),
        to_int(row.get('driver_age')),
        row.get('driver_race'),
        row.get('violation'),
        to_bool(row.get('search_conducted')),
        row.get('search_type'),
        row.get('stop_outcome'),
        to_bool(row.get('drugs_related_stop')),
        row.get('stop_duration'),
        row.get('vehicle_number') or f"VEH{idx}",
        row.get('timestamp')
    )

    cursor.execute(insert_query, values)

connection.commit()
print("✅ Data inserted successfully")
```
